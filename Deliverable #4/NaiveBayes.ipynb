# Naive Bayes

### Concepts Learned:
1. Naive Bayes
    * A Classification Probablistic Machine Learning Model
    * It is only capable of classification and can't produce contious numbers.
    * It is Naive as it does consider order of data
2. Multinomial NB
    * A Version of Naive Bayes that uses only disceerte data.
    * Steps:
        * Calcuate Prior Probability: For each Class count the instance of each class and divided it by number of all instances.
        * Calcuate Likelihood: For each class, calcualte all P(X|C), where X is each feature. Its best to have an Alpha of atleast one in this step. 
        * Afterwards you can begin with prediction.
        * Enter a input. Count the instance ofe each feature in the input. For each class, calcuate the log of its Prior Probability *  Likelihood of each instance that appeared. The class that has highest value is prediction.
3. Gaussian
    * A Version of Naive Bayes that uses only numerical data.
    * Steps:
        * For each feature, create a Gaussian curve by finding each feature's variation and mean.
        * Calcuate Prior Probability: For each Class count the instance of each class and divided it by number of all instances.
        * Unlike Multinomial NB, Likelihood is not calcuated in the training phase. Instead the Gaussian Curves will be used to find likelihood of each feature in the inference phase.
        *  * Enter a input. Count the instance ofe each feature in the input. For each class, calcuate the log of its Prior Probability *  Likelihood of each instance that appeared(using the Gausian Curve). The class that has highest value is prediction.

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.naive_bayes import MultinomialNB, GaussianNB
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.datasets import load_breast_cancer
from sklearn.metrics import f1_score, accuracy_score

import pandas as pd
import numpy as np
```

## MultinomialNB Model

```python
data = pd.read_csv("df_file.csv")
data.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(data['Text'], data['Label'], shuffle=True, test_size=0.2)
```

```python
pipeline = Pipeline([
    ("Vectorizer", TfidfVectorizer()),
    ("Model", MultinomialNB(alpha=1.0))
])
```

```python
pipeline.fit(X_train, y_train)
```

```python
y_pred = pipeline.predict(X_test)
f = f1_score(y_true=y_test, y_pred=y_pred, average="macro")
a = accuracy_score(y_true=y_test, y_pred=y_pred)


print(f"F1 Score: {f}")
print(f"Accuracy: {a*100}%")
```

The Main Problem with Naive Bayes is that it assume everything is independent. This causes problems when your datasets contains indpendent varibales with high correlation.

Say you have a dataset that has "Body Temperature" and "Fever Likelness". Those 2 features are high correlated and represent the same signal. However, an NB Model doesn't know with that, and treat both of those featues are completely independent. This leads to the model double counting that singal, giving the signal more weight and effect in the model, than it should.

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
data = load_iris(as_frame=True).frame
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
sns.histplot(data=data, x="sepal length (cm)")
plt.show()
```

```python
data["Norm"] = MinMaxScaler().fit_transform(data[["sepal length (cm)"]]) 
sns.histplot(data=data, x="Norm")
plt.show()
```

```python
data["scale"] = StandardScaler().fit_transform(data[["sepal length (cm)"]]) 
sns.histplot(data=data, x="scale")
plt.show()
```
